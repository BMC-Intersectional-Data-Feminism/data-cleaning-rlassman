```python
import pandas as pd 
import numpy as np
```

```python
df_example = pd.read_csv('audible_uncleaned.csv')
df_example
```

```python
## AUTHOR

df_example['author'] = df_example['author'].str.lstrip('Writtenby:')

df_example
```

```python
import matplotlib.pyplot as plt ##package to do visualization

valueCounts = df_example['author'].head(100).value_counts() ##.head(100) = first 100, counting up times author appears in first 100 lines

valueCounts.plot(kind='bar') ##make bar graph
```

```python
## NARRATOR

df_example['narrator'] = df_example['narrator'].str.lstrip('Narratedby:')

df_example
```

```python
## TIME
def convertTime(timeString):
    totalMins = 0
    splitString = timeString.split(' and ')
    for item in splitString:
        if 'hrs' in item:
            hours = int(item.split(' ')[0])
            totalMins += (60*hours)
        elif 'mins' in item:
            mins = int(item.split(' ')[0])
            totalMins += mins
    return totalMins

df_example['time'] = df_example['time'].apply(convertTime)
df_example
```

```python
## RELEASE DATE

df_example['releasedate'] = pd.to_datetime(df_example["releasedate"], infer_datetime_format=True, format='mixed').dt.strftime('%Y-%m-%d')
df_example
```

```python
## STARS

notRated = df_example[df_example['stars'] == 'Not rated yet'] ## select rows with no ratings
rated = df_example[df_example['stars'] != 'Not rated yet'] ## rows with ratings

rated
```

```python
rated[['stars', 'numRatings']] = rated['stars'].str.split('stars', expand = True) ## expand -> create 2 new columns
rated
```

```python
rated['stars'] = rated['stars'].str.split(' out of ').str[0]

df_cleaned = pd.concat(rated, notRated)
```
